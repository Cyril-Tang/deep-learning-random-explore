# Basic

```python
learn = fake_learner(n_in=2, n_out=2, batch_size=2, train_length=20)
```

```python
learn.fit_one_cycle(6, 0.1)
```

```python
learn.recorder.plot_lr(True)
```

```python
lr1, mom1 = learn.recorder.lrs, learn.recorder.moms
```

```python
learn = fake_learner(n_in=2, n_out=2, batch_size=2, train_length=20)
```

```python
learn.fit_one_cycle(2, 0.1, tot_epochs=6)
```

```python
learn.recorder.plot_lr(True)
```

```python
lr2, mom2 = learn.recorder.lrs, learn.recorder.moms
```

```python
learn.save('epoch2')
```

```python
learn.load('epoch2')
```

```python
learn.fit_one_cycle(4, 0.1, start_epoch=3, tot_epochs=6)
```

```python
learn.recorder.plot_lr(True)
```

```python
lr3, mom3 = learn.recorder.lrs, learn.recorder.moms
```

```python
lr1 == lr2 + lr3, mom1 == mom2 + mom3
```
